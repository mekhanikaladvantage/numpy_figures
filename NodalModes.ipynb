```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
#we need to compute the zeroes of the Bessel function so our graph scales properly
#specifically to make the figure I need the first three.
#luckily scipy makes this very easy.
lambdas = sp.jn_zeros(0,3)
lambdas
```

```python
r = np.linspace(0,15,1000)
plt.plot (r, sp.j0(lambdas[2]*r))
plt.plot(r,np.zeros(1000),'k')
plt.plot(np.zeros(1000),r,'k')
plt.ylim((-0.5,1.1))
plt.xlim((0,1))
plt.savefig("modal_node3.png")
```
