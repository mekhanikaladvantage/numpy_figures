```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
#we need to compute the zeroes of the Bessel function so our graph scales properly
#specifically to make the figure I need the first three.
#luckily scipy makes this very easy.
lambdas = sp.jn_zeros(0,3)
lambdas
```

```python
r = np.linspace(0,15,1000)
plt.plot (r, sp.j0(lambdas[2]*r))
plt.plot(r,np.zeros(1000),'k')
plt.plot(np.zeros(1000),r,'k')
plt.ylim((-0.5,1.1))
plt.xlim((0,1))
plt.savefig("modal_node3.png")
```

```python
#Next time we do the fun part and make a 3d mesh, which as of 7/20 I have zero clue how to accomplish. 
#Essentially what I need is a radial shell, as in I take the plot of J_0(lambda*r) and rotate it 2pi radians
#This will trace out a surface
#First thing I neeed is to set up 3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
#now lets make the mesh
r = np.linspace(0,1,100)
theta = np.linspace(0,2*np.pi,100)
R,Theta = np.meshgrid(r,theta)
#map will just be sp.j0(R), scaled by the proper lambda
#convert to cartesian
X,Y = R*np.cos(Theta),R*np.sin(Theta)

#now plot
ax.plot_surface(X,Y,sp.j0(lambdas[2]*R), cmap=plt.cm.Blues_r)
ax.set_zlim(-0.5,2)
ax.grid(None)
ax.set_axis_off()
plt.savefig('nodal_modes_3d_2.png')

```
