```python
def phi(x):
    if x > -2 and x <= 0:
        return 0
    elif x >= 0 and x < 2:
        return x
    else:
        pass

#ok lets attempt figure 78
#define coefficients, which I have calculated exactly
def a(n):
    if (n % 2) == 0:
        return 0
    else:
        return (-4/(n**2*(np.pi)**2))
    
def b(n):
    return ((-1)**(n+1)*2)/(n*np.pi)

#coeffs seem to be ok
#now figure out sum
def f(x, n):
    terms = np.array([a(i)*np.cos((i*np.pi*x)/2)+b(i)*np.sin((i*np.pi*x)/2) for i in range(1,n+1)])
    return terms.sum() + 1/2

xvals = np.arange(-2,2,0.01)
yvals1 = np.array([f(x,1) for x in xvals])
yvals4 = np.array([f(x,4) for x in xvals])
yvals6 = np.array([f(x,6) for x in xvals])
yvals16 = np.array([f(x,16) for x in xvals])
yvals21 = np.array([f(x,21) for x in xvals])
yvals_phi = np.array([phi(x) for x in xvals])

plt.plot(xvals,yvals_phi, '--')
plt.plot(xvals,yvals21,"b-")
plt.grid()
plt.title("N=21")

plt.savefig('fig78_n21.png')

#lets figure out subplots
# fig, axs = plt.subplots(2,3)

# axs[0,0].plot(xvals,yvals_phi, '--')
# axs[0,0].plot(xvals,yvals1, 'b-')
# axs[0,0].set_title("N=1")

# axs[0,1].plot(xvals,yvals_phi, '--')
# axs[0,1].plot(xvals,yvals4, 'b-')
# axs[0,1].set_title("N=4")

# axs[0,2].plot(xvals,yvals_phi, '--')
# axs[0,2].plot(xvals,yvals6, 'b-')
# axs[0,2].set_title("N=6")

# axs[1,0].plot(xvals,yvals_phi, '--')
# axs[1,0].plot(xvals,yvals16, 'b-')
# axs[1,0].set_title("N=16")

# axs[1,1].plot(xvals,yvals_phi, '--')
# axs[1,1].plot(xvals,yvals21, 'b-')
# axs[1,1].set_title("N=21")
```
